```python
def assign_year_interval(year):
    try:
        year = int(year)
        if 1890 <= year <= 1917:
            return '1890-1917'
        elif 1917 < year <= 1945:
            return '1917-1945'
        elif 1945 < year <= 1970:
            return '1945-1970'
        elif 1970 < year <= 1990:
            return '1970-1990'
        elif 1991 <= year <= 2025:
            return '1991-2025'
        else:
            return 'не попал'
    except:
        return 'не попал'

df['year_built_interval'] = df['info_year_built'].apply(assign_year_interval)
```

```python
df = df.drop(columns=['info_year_built'])
```

```python
df = df.dropna()

df = df.drop_duplicates()

print(df.info())
```

```python
def haversine(coords):
    R = 6371
    lat = np.radians(coords[:, 0])
    lon = np.radians(coords[:, 1])

    lat1 = lat[:, np.newaxis]
    lon1 = lon[:, np.newaxis]
    lat2 = lat[np.newaxis, :]
    lon2 = lon[np.newaxis, :]

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    return R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
```
